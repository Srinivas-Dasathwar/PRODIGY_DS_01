```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_580248.csv", index_col=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'])
```

```python
df
```

```python
countries = ["United States", "China", "India"]
population_data = df.loc[countries, :]
```

```python
years = population_data.columns[4:]
population_values = population_data.iloc[:, 4:]
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 1, 1)
population_values.T.plot(kind='bar', ax=plt.gca())
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Bar Chart Comparison')
plt.legend(title='Country')
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1, 2, 2)
population_values.T.plot(kind='hist', bins=10, alpha=0.7, stacked=True, ax=plt.gca())
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Histogram Comparison')
plt.legend(title='Country')
```
